```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


from sklearn.datasets import load_boston
```

```python
bh = load_boston()
```

```python
df = pd.DataFrame(bh.data, columns = bh.feature_names)
df['MEDV'] = bh.target

```

```python
x = df.CRIM
y = df.MEDV

x = np.log(x)

coef = np.polyfit(x,y, deg=1)
x_values = np.linspace(x.min(),x.max(),num=10)
y_values = np.polyval(coef,x_values)


plt.scatter(x, y, s=15)
plt.plot(x_values,y_values,label='Linear regression', c= 'r')

plt.xlabel('Criminality')
plt.ylabel('Median house price')

plt.legend()
plt.show()

```

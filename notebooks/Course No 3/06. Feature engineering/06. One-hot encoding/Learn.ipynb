```python
import pandas as pd
import numpy as np
```

```python
data_df = pd.read_csv('bike-small.csv')
```

```python
data_df.head()
```

```python
data_df.weekday.value_counts()
```

```python
#R2 score with temperature only

X = data_df[['temp']].values
y = data_df['casual'].values

from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(X,y)

print("R2 score without temperature only: {:.03f}".format(lr.score(X,y)))
```

```python
#R2 score without modification on weekdays

X = data_df.drop('casual',axis=1).values
y = data_df['casual'].values

from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(X,y)

print("R2 score without modifying weekday numbers: {:.03f}".format(lr.score(X,y)))
```

```python
# Use the get_dummies function from Pandas


encoded_df = pd.get_dummies(data_df,columns=['weekday'])
encoded_df.head()
```

```python
#R2 score without modification on weekdays

X = encoded_df.drop('casual',axis=1).values
y = encoded_df['casual'].values


lr = LinearRegression()

lr.fit(X,y)

print("R2 score without modifying weekday numbers: {:.03f}".format(lr.score(X,y)))
```

```python
# Drop first column


encoded_df = pd.get_dummies(data_df,columns=['weekday'], drop_first=True)
encoded_df.head()
```

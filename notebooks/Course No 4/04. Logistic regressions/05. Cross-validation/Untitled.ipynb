```python
from sklearn import datasets

# Load data set
iris = datasets.load_iris()

# Create X/y arrays
X = iris['data']
y = iris['target']
l = iris['target_names']
```

```python
from sklearn.model_selection import KFold

# Create k-fold object
kfold = KFold(n_splits=3)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.pipeline import Pipeline

# Create a logistic regression
#logreg = make_pipeline(
#    StandardScaler(), LogisticRegression(C=1, solver='liblinear', multi_class='ovr'))

logreg=Pipeline([
    ('scaler', StandardScaler()),
    ('knn', LogisticRegression(C=1000, solver='liblinear', multi_class='ovr', max_iter=10000))
])
```

```python
################### XAVIER approach with GridSearchCV ###########

from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import ParameterGrid
import pandas as pd

parameters = {'knn__C':[1, 10, 100, 1000], 'knn__solver':['sag', 'saga', 'lbfgs'], 'knn__multi_class':['ovr','multinomial']}
grid=GridSearchCV(logreg, parameters, n_jobs=-1, cv=10, return_train_score=True, iid=True)

grid.get_params().keys()

grid.fit(X, y)

pd.DataFrame(grid.cv_results_)[['param_knn__C', 'param_knn__solver', 'param_knn__multi_class', 'mean_test_score', 'mean_train_score']].sort_values('mean_test_score', ascending=False)
```

```python
grid.score(X, y)
```

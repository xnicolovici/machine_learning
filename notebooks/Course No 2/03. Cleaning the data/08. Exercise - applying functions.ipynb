 * N = Nabisco

1. Calories can be converted to kilojoules using the following formula: 1 calorie = 4.184 kilojoules. Find for the amount of Kilojoules per serving for each cereal. Store the results in a new column.

1. For each numerical column, find the cereal that has the maximum value in that column.

```python
mfr_map ={
    'A': 'American Home Fodd Products',
    'G': 'General Mills',
    'K': 'Kelloggs',
    'N': 'Nabisco',
    'P': 'Post',
    'Q': 'Quaker Oats',
    'R': 'Ralston Purina'
}
```

```python
df['mfr full name'] = df.mfr.map(mfr_map)
```

```python
df['kilojoule'] = df.calories.apply(lambda x: x * 4.184)
```

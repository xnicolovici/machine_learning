```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('financial_data.csv', index_col='Date')
data.index = pd.to_datetime(data.index) # Convert the index
prices = data['Close']
```

```python
prices.rolling(260,center=True).mean().dropna().plot()
```

```python
# Create the figure
plt.figure(figsize=(8, 4))

# Compute rolling mean
rolling_mean_1year = prices.rolling(260, center=True).mean()

# Plot the results
prices.plot(label='prices', alpha=0.5)
rolling_mean_1year.plot(label='rolling mean')
plt.legend()
plt.show()
```

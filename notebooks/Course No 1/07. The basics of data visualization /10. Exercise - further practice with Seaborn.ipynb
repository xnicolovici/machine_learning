```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

iris = sns.load_dataset('iris')
```

```python
iris.head(10)
```

```python
sns.violinplot(x=iris.species, y=iris.petal_length, palette='pastel');
```

```python
sns.swarmplot(x=iris.species, y=iris.sepal_length, palette='pastel');
```

```python
sns.swarmplot('species', 'sepal_length', data=iris, palette='pastel');
```

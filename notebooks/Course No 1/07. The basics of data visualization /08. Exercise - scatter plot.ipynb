```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import datasets
iris = datasets.load_iris()
wine = datasets.load_wine()
```

```python

iris_df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
iris_df['species'] = iris['target']
```

```python

c = ['red', 'yellow', 'green']

for i in range(3):
    df = iris_df[iris_df['species'] == i]
    plt.scatter(df[iris_df.columns[0]], df[iris_df.columns[2]], color=c[i], label='Iris ' + iris['target_names'][i])

    plt.xlabel(iris_df.columns[2])
plt.ylabel(iris_df.columns[0])

plt.legend(loc='lower right')
plt.title('Iris dataset: ' + iris_df.columns[2] + ' vs ' + iris_df.columns[0])

plt.show()
```

```python
music = pd.read_csv('../../../data/songs_requested.csv')
```

```python
music.groupby('Musician')['Requested'].describe()
```

```python
music.groupby('Musician')['Decade'].agg(lambda x: str(min(x))+"'s - "+str(max(x))+"'s").to_frame()
```

```python
grouped_df=music.groupby('Musician').agg({'Name': 'count',
                                    'Decade': lambda x: str(min(x))+"-"+str(max(x)), 
                                    'Requested': ['sum', 'max', np.mean]}
                                    )
grouped_df
```

```python
grouped_df.rename(columns={"count":"Total",
                           "<lambda>":"span"}, inplace=True)
```

```python
grouped_df.to_csv('export.csv')
```
